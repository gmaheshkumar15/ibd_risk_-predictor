```python
# ======= Step 0: Import Libraries =======
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# ======= Step 1: Prepare Data =======
X = data.drop(columns=['Label'])
y = data['Label']

# ======= Step 3: Apply PCA =======
n_components = 10
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)

# Convert to DataFrame
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(n_components)])
pca_df['Group'] = y

# Explained variance
explained_var = pca.explained_variance_ratio_
print("Explained variance ratio for 10 PCs:\n", explained_var)
print("Cumulative explained variance:", explained_var.cumsum())

# ======= Step 4: Loadings / Dietary Patterns =======
loadings = pd.DataFrame(pca.components_.T, index=X.columns, columns=[f'PC{i+1}' for i in range(n_components)])

# Classify contribution (optional)
threshold = 0.3
pattern_dict = {}
for i in range(n_components):
    pc = f'PC{i+1}'
    pattern_dict[pc] = loadings[pc].apply(lambda x: 'Positive' if x>threshold else ('Negative' if x<-threshold else 'Neutral'))

pattern_df = pd.DataFrame(pattern_dict)

# ======= Step 5: Top Contributing Foods =======
top_foods_summary = {}
for i in range(n_components):
    pc = f'PC{i+1}'
    top_positive = loadings[pc][loadings[pc] > threshold].sort_values(ascending=False).index.tolist()
    top_negative = loadings[pc][loadings[pc] < -threshold].sort_values().index.tolist()
    top_foods_summary[pc] = {'Positive foods': top_positive, 'Negative foods': top_negative}

# Print summary
for i, pc in enumerate(top_foods_summary.keys()):
    print(f"\n{pc} (Variance explained: {explained_var[i]:.2f})")
    print("  Positive foods:", top_foods_summary[pc]['Positive foods'])
    print("  Negative foods:", top_foods_summary[pc]['Negative foods'])

# ======= Step 6: Compare PC Scores by Group =======
pc_group = pca_df.groupby('Group')[[f'PC{i+1}' for i in range(n_components)]].mean()
print("\nMean PC scores by group:\n", pc_group)

# ======= Step 7: PCA Scatter Plot (PC1 vs PC2) =======
plt.figure(figsize=(12,8))
sns.set(style="whitegrid")
sns.scatterplot(x='PC1', y='PC2', hue='Group', data=pca_df, palette={0:'blue',1:'red'}, s=100)
plt.title('PCA: PC1 vs PC2 (IBD vs Controls)')
plt.xlabel('PC1')
plt.ylabel('PC2')

# Overlay top contributing foods
top_foods_pc1_pc2 = loadings[(loadings['PC1'].abs() > threshold) | (loadings['PC2'].abs() > threshold)]
for i, row in top_foods_pc1_pc2.iterrows():
    plt.text(row['PC1']*5, row['PC2']*5, i, fontsize=9, color='black')

plt.show()

# ======= Step 8: Explained Variance Plot =======
plt.figure(figsize=(10,6))
plt.bar(range(1,n_components+1), explained_var*100, alpha=0.7, color='teal', label='Individual')
plt.step(range(1,n_components+1), explained_var.cumsum()*100, where='mid', label='Cumulative', color='orange')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance (%)')
plt.title('Explained Variance by Principal Components')
plt.legend()
plt.show()
```

```python
# ======= Step 9: Visualize 10 Dietary Patterns =======
n_top = 3  # top 3 positive and negative foods per PC
plt.figure(figsize=(18, 14))

for i in range(n_components):
    pc = f'PC{i+1}'
    explained = explained_var[i]*100

    # Get top foods
    top_pos = loadings[pc].sort_values(ascending=False).head(n_top)
    top_neg = loadings[pc].sort_values().head(n_top)

    # Combine for plotting
    foods = pd.concat([top_pos, top_neg])
    colors = ['green']*n_top + ['red']*n_top

    # Subplot
    plt.subplot(5, 2, i+1)  # 5 rows, 2 columns
    plt.barh(foods.index, foods.values, color=colors)
    plt.title(f'{pc} (Explained Variance: {explained:.1f}%)')
    plt.xlabel('Loading')
    plt.gca().invert_yaxis()  # largest on top
    plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# ======= Step 0: Import Libraries =======
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# ======= Step 1: Prepare Data =======
# Assume your data is in a DataFrame called `data` with "Label" as group column (0=Controls, 1=IBD)
X = data.drop(columns=['Label'])
y = data['Label']

# Standardize the dietary intake data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ======= Step 2: Apply PCA =======
n_components = 10
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)

# Convert to DataFrame
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(n_components)])
pca_df['Group'] = y

# Explained variance
explained_var = pca.explained_variance_ratio_
print("Explained variance ratio for 10 PCs:\n", explained_var)
print("Cumulative explained variance:", explained_var.cumsum())

# ======= Step 3: Loadings / Dietary Patterns =======
loadings = pd.DataFrame(
    pca.components_.T,
    index=X.columns,
    columns=[f'PC{i+1}' for i in range(n_components)]
)

# Classify contribution (optional)
threshold = 0.1
pattern_dict = {}
for i in range(n_components):
    pc = f'PC{i+1}'
    pattern_dict[pc] = loadings[pc].apply(
        lambda x: 'Positive' if x > threshold else ('Negative' if x < -threshold else 'Neutral')
    )
pattern_df = pd.DataFrame(pattern_dict)

# ======= Step 4: Top Contributing Foods =======
top_foods_summary = {}
for i in range(n_components):
    pc = f'PC{i+1}'
    top_positive = loadings[pc][loadings[pc] > threshold].sort_values(ascending=False).index.tolist()
    top_negative = loadings[pc][loadings[pc] < -threshold].sort_values().index.tolist()
    top_foods_summary[pc] = {'Positive foods': top_positive, 'Negative foods': top_negative}

# Print summary
for i, pc in enumerate(top_foods_summary.keys()):
    print(f"\n{pc} (Variance explained: {explained_var[i]:.2f})")
    print("  Positive foods:", top_foods_summary[pc]['Positive foods'])
    print("  Negative foods:", top_foods_summary[pc]['Negative foods'])

# ======= Step 5: Compare PC Scores by Group =======
pc_group = pca_df.groupby('Group')[[f'PC{i+1}' for i in range(n_components)]].mean()
print("\nMean PC scores by group:\n", pc_group)

# ======= Step 6: Interpretation (Controls vs IBD) =======
for i in range(n_components):
    pc = f'PC{i+1}'
    control_mean = pc_group.loc[0, pc]  # 0=Controls
    ibd_mean = pc_group.loc[1, pc]      # 1=IBD
    
    if control_mean > ibd_mean:
        print(f"\n{pc}: Controls have higher scores")
        print("  Positive foods -> linked to Controls")
        print("  Negative foods -> linked to IBD")
    else:
        print(f"\n{pc}: IBD has higher scores")
        print("  Positive foods -> linked to IBD")
        print("  Negative foods -> linked to Controls")

# ======= Step 7: PCA Scatter Plot (PC1 vs PC2) =======
plt.figure(figsize=(12,8))
sns.set(style="whitegrid")
sns.scatterplot(x='PC1', y='PC2', hue='Group', data=pca_df,
                palette={0:'blue',1:'red'}, s=100)
plt.title('PCA: PC1 vs PC2 (IBD vs Controls)')
plt.xlabel('PC1')
plt.ylabel('PC2')

# Overlay top contributing foods
top_foods_pc1_pc2 = loadings[(loadings['PC1'].abs() > threshold) | (loadings['PC2'].abs() > threshold)]
for i, row in top_foods_pc1_pc2.iterrows():
    plt.text(row['PC1']*5, row['PC2']*5, i, fontsize=9, color='black')

plt.show()

# ======= Step 8: Explained Variance Plot =======
plt.figure(figsize=(10,6))
plt.bar(range(1,n_components+1), explained_var*100, alpha=0.7, color='teal', label='Individual')
plt.step(range(1,n_components+1), explained_var.cumsum()*100, where='mid',
         label='Cumulative', color='orange')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance (%)')
plt.title('Explained Variance by Principal Components')
plt.legend()
plt.show()

# ======= Step 9: Visualize 10 Dietary Patterns =======
n_top = 3  # top 3 positive and negative foods per PC
plt.figure(figsize=(18, 14))

for i in range(n_components):
    pc = f'PC{i+1}'
    explained = explained_var[i]*100

    # Get top foods
    top_pos = loadings[pc].sort_values(ascending=False).head(n_top)
    top_neg = loadings[pc].sort_values().head(n_top)

    # Combine for plotting
    foods = pd.concat([top_pos, top_neg])
    colors = ['green']*n_top + ['red']*n_top

    # Subplot
    plt.subplot(5, 2, i+1)  # 5 rows, 2 columns
    plt.barh(foods.index, foods.values, color=colors)
    plt.title(f'{pc} (Explained Variance: {explained:.1f}%)')
    plt.xlabel('Loading')
    plt.gca().invert_yaxis()
    plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# ======= Step 10: Heatmap of Loadings =======
plt.figure(figsize=(14,10))
sns.heatmap(
    loadings, 
    cmap="coolwarm", 
    center=0, 
    annot=True,            # Show the loading values
    fmt=".2f",             # Format numbers to 2 decimal places
    linewidths=0.5,        # Optional: lines between cells
    cbar_kws={'label': 'Loading'}
)
plt.title("Food Loadings on 10 Principal Components")
plt.ylabel("Food Items")
plt.xlabel("Principal Components")
plt.tight_layout()
plt.show()


# ======= Step 11: Variance Summary Table =======
variance_df = pd.DataFrame({
    "PC": [f"PC{i+1}" for i in range(n_components)],
    "Explained Variance (%)": explained_var*100,
    "Cumulative (%)": explained_var.cumsum()*100
})
print("\nVariance Summary:\n", variance_df)

```

```python
# ======= Step 13: Auto-Naming Dietary Patterns =======
def name_pattern(loadings, n_top=3):
    """Generate a name for each PC using top positive and negative foods."""
    names = {}
    for pc in loadings.columns:
        top_pos = loadings[pc].sort_values(ascending=False).head(n_top).index.tolist()
        top_neg = loadings[pc].sort_values().head(n_top).index.tolist()

        # Build name
        pos_name = ", ".join(top_pos)
        neg_name = ", ".join(top_neg)
        names[pc] = f"High in [{pos_name}] vs Low in [{neg_name}]"
    return names

pattern_names = name_pattern(loadings, n_top=3)

print("\n===== Dietary Pattern Names =====")
for pc, name in pattern_names.items():
    print(f"{pc}: {name}")
```

```python
from scipy.stats import ttest_ind

for pc in [f'PC{i+1}' for i in range(10)]:
    controls = pca_df[pca_df['Group']==0][pc]
    ibd = pca_df[pca_df['Group']==1][pc]
    stat, pval = ttest_ind(controls, ibd)
    print(f"{pc}: p={pval:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate mean PC scores by group
pc_group_means = pca_df.groupby('Group')[[f'PC{i+1}' for i in range(10)]].mean()

# Determine linkage for each PC
pc_link = {}
for pc in pc_group_means.columns:
    if pc_group_means.loc[1, pc] > pc_group_means.loc[0, pc]:  # 1=IBD, 0=Controls
        pc_link[pc] = 'IBD'
    else:
        pc_link[pc] = 'Controls'

# Create colors for PCs based on linkage
pc_colors = ['#FF9999' if pc_link[pc]=='IBD' else '#99CCFF' for pc in loadings.columns]

# Mask small loadings for clarity
threshold = 0.15
loadings_masked = loadings.mask(loadings.abs() < threshold)

plt.figure(figsize=(14,10))
sns.heatmap(
    loadings_masked, 
    cmap="coolwarm", 
    center=0, 
    annot=True, 
    fmt=".2f", 
    linewidths=0.5, 
    cbar_kws={'label': 'Loading'},
    yticklabels=True
)

# Color PC labels
for tick_label, color in zip(plt.gca().get_xticklabels(), pc_colors):
    tick_label.set_color(color)
    tick_label.set_fontweight('bold')

plt.title("Top Food Loadings on 10 Principal Components (Red=IBD, Blue=Controls)")
plt.ylabel("Food Items")
plt.xlabel("Principal Components")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
