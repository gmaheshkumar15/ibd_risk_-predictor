```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu, ks_2samp, entropy


# =========================
# Prepare IBD column (UC + CD)
# =========================
data['IBD'] = data['DIAGNOSIS'].apply(lambda x: 'IBD' if x in ['UC','CD'] else 'Control')

# =========================
# Define features and comparisons
# =========================
features = [col for col in data.columns if col not in ['DIAGNOSIS', 'IBD']]
comparisons = [
    ('CONTROLS', 'IBD'),
    ('CONTROLS', 'UC'),
    ('CONTROLS', 'CD'),
    ('UC', 'CD')
]

epsilon = 1e-10  # for KL divergence
results = []

# =========================
# Helper functions
# =========================
def normalize(dist):
    dist = np.array(dist, dtype=float) + epsilon
    return dist / np.sum(dist)

# =========================
# Run tests for each feature
# =========================
for feature in features:
    for grp1, grp2 in comparisons:
        # Select values for each group
        vals1 = data.loc[data['DIAGNOSIS'].isin([grp1]) if grp1 not in ['IBD'] else data['IBD']=='IBD', feature].dropna().values
        vals2 = data.loc[data['DIAGNOSIS'].isin([grp2]) if grp2 not in ['IBD'] else data['IBD']=='IBD', feature].dropna().values

        # Mann-Whitney U test
        try:
            u_stat, p_mw = mannwhitneyu(vals1, vals2, alternative='two-sided')
        except:
            p_mw = np.nan

        # KS test
        try:
            ks_stat, p_ks = ks_2samp(vals1, vals2)
        except:
            ks_stat = ks_stat = np.nan

        # KL divergence
        # Convert to histogram-based PDF
        bins = np.histogram_bin_edges(np.concatenate([vals1, vals2]), bins='auto')
        hist1, _ = np.histogram(vals1, bins=bins, density=True)
        hist2, _ = np.histogram(vals2, bins=bins, density=True)
        hist1 += epsilon
        hist2 += epsilon
        kl1 = entropy(hist1, hist2)
        kl2 = entropy(hist2, hist1)
        kl_sym = 0.5 * (kl1 + kl2)

        # Store results
        results.append({
            'Feature': feature,
            'Comparison': f'{grp1} vs {grp2}',
            'Mann-Whitney U test p value': p_mw,
            'KS test p value': p_ks,
            'Relative entropy': kl_sym
        })

# =========================
# Convert results to DataFrame
# =========================
results_df = pd.DataFrame(results)
pd.set_option('display.float_format', '{:.5f}'.format)

# =========================
# Print results
# =========================
print(results_df)

# =========================
results_df.to_excel('pairwise_tests_results.xlsx', index=False)
```
