```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_excel('Merged.xlsx')
```

```python
data.head()
```

```python
data.shape
```

IBD

```python
ibd=data[data['Label']==1]
```

```python
ibd
```

```python
# Preprocessing (optional but recommended)
scaler = StandardScaler()
ibd_scaled = scaler.fit_transform(ibd)

# Step 4: Choose number of clusters (k) — Elbow method
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(ibd_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Step 4: Choose number of clusters (k) — Elbow method
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(ibd)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Step 5: Fit KMeans with chosen k
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(ibd)

# Step 6: Get cluster labels and add to dataframe
ibd['Cluster'] = kmeans.labels_
```

```python
import pandas as pd

# Assuming 'ibd' is your dataframe with the 'Cluster' column
output_file = "ibd_clusters.xlsx"

# Export to Excel
ibd.to_excel(output_file, index=False)

print(f"Clustered data saved to {output_file}")
```

```python
ibd.head()
```

```python
ibd['Cluster'].value_counts()
```

```python
ibd.groupby('Cluster').agg(['mean','std'])
```

```python
# Compute grouped statistics
cluster_stats = ibd.groupby('Cluster').agg(['mean', 'std'])

# Save to Excel
cluster_stats.to_excel("ibd_cluster_stats.xlsx")

print("Cluster statistics saved to ibd_cluster_stats.xlsx")
```

```python
from sklearn.metrics import silhouette_score

# Step 7: Calculate Silhouette Score
score = silhouette_score(ibd.drop('Cluster', axis=1), ibd['Cluster'])
print(f"Silhouette Score: {score:.3f}")
```
