```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_excel('Merged.xlsx')
```

```python
X = data.drop(columns=['Label'])  # All diet features
y = data['Label']                 # Group labels (0 or 1)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10, random_state=42)
X_pca = pca.fit_transform(X)
```

```python
explained_var = pca.explained_variance_ratio_
print("Variance explained by each pattern:", explained_var)
print("Total variance explained:", explained_var.sum())
```

```python
 #3. Create DataFrame for PCA scores (diet patterns)
diet_patterns = pd.DataFrame(
    X_pca,
    columns=[f'Pattern_{i+1}' for i in range(10)]
)

# 4. Loadings (feature contributions to each pattern)
loadings = pd.DataFrame(
    pca.components_,
    columns=X.columns,   # works if X is a DataFrame
    index=[f'Pattern_{i+1}' for i in range(10)]
)
```

```python
import pandas as pd

# Split the text into lines
lines = loadings_text.strip().split('\n')

# Prepare lists
foods = []
loadings = []

for line in lines:
    # Split each line into words/numbers
    parts = line.split()
    food_name = " ".join(parts[:-10])  # Everything except the last 10 numbers
    numbers = list(map(float, parts[-10:]))  # Last 10 entries are the PC loadings
    
    foods.append(food_name)
    loadings.append(numbers)

# Create DataFrame
loadings_df = pd.DataFrame(loadings, index=foods, columns=[f'PC{i+1}' for i in range(10)])

# View
print(loadings_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Suppose you already have this dictionary from your previous analysis:
# Key: PC, Value: 'IBD' or 'Controls' based on mean PC scores
pc_link = {
    'PC1': 'IBD', 'PC2': 'Controls', 'PC3': 'Controls', 'PC4': 'IBD',
    'PC5': 'Controls', 'PC6': 'Controls', 'PC7': 'IBD', 'PC8': 'IBD',
    'PC9': 'IBD', 'PC10': 'IBD'
}

# Create colors for PCs based on linkage
pc_colors = ['#FF9999' if pc_link[pc]=='IBD' else '#99CCFF' for pc in loadings.columns]

# Mask small loadings
threshold = 0.2
loadings_masked = loadings.mask(loadings.abs() < threshold)

plt.figure(figsize=(14,10))
sns.heatmap(
    loadings_masked, 
    cmap="coolwarm", 
    center=0, 
    annot=True, 
    fmt=".2f", 
    linewidths=0.5, 
    cbar_kws={'label': 'Loading'},
    yticklabels=True
)

# Add colored labels on top for IBD vs Controls
for tick_label, color in zip(plt.gca().get_xticklabels(), pc_colors):
    tick_label.set_color(color)
    tick_label.set_fontweight('bold')

plt.title("Top Food Loadings on 10 Principal Components (Red=IBD, Blue=Controls)")
plt.ylabel("Food Items")
plt.xlabel("Principal Components")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
